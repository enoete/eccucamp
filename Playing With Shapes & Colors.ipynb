```python
# Assistant
# Import the required libraries
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt

# Create a new canvas for shapes demo
shapes_canvas = Image.new('RGB', (600, 400), 'lightblue')
shapes_draw = ImageDraw.Draw(shapes_canvas)

# 1. RECTANGLES - defined by two corners
print("üî≤ Drawing rectangles...")
shapes_draw.rectangle([50, 50, 150, 100], fill='red', outline='black', width=2)
shapes_draw.rectangle([200, 50, 300, 100], fill='green')

# 2. CIRCLES/ELLIPSES - defined by bounding box
print("üîµ Drawing circles...")
shapes_draw.ellipse([350, 50, 450, 150], fill='blue', outline='navy', width=3)
shapes_draw.ellipse([500, 50, 550, 100], fill='yellow')  # Oval

# 3. LINES - defined by start and end points
print("üìè Drawing lines...")
shapes_draw.line([(50, 200), (150, 250)], fill='purple', width=5)
shapes_draw.line([(200, 200), (300, 200), (250, 250), (200, 200)], fill='orange', width=3)  # Connected lines

# 4. POLYGONS - defined by list of points
print("üî∫ Drawing polygons...")
triangle_points = [(400, 200), (450, 150), (500, 200)]
shapes_draw.polygon(triangle_points, fill='pink', outline='red', width=2)

# Star shape
star_points = [(350, 300), (365, 270), (380, 300), (360, 285), (370, 315)]
shapes_draw.polygon(star_points, fill='gold', outline='orange', width=2)

# Display shapes demo
plt.figure(figsize=(12, 8))
plt.imshow(shapes_canvas)
plt.title("üé® Basic Shapes Toolkit")
plt.axis('off')
plt.show()

print("‚úÖ Basic shapes mastered!")
print("üí° Key concept: All shapes are defined by coordinates (x, y)")
print("   ‚Ä¢ (0, 0) is top-left corner")
print("   ‚Ä¢ x increases going right")
print("   ‚Ä¢ y increases going down")
```

## üá∞üá≥ Quest 1: The National Flag (Easy Level)

**Your Mission:** Recreate the flag of St. Kitts and Nevis using geometric shapes.

### üéØ Flag Design Elements:
- **Two green triangles:** Upper left and lower right
- **Two red triangles:** Upper right and lower left  
- **Black diagonal band:** Running from lower left to upper right
- **Two yellow stripes:** Bordering the black band
- **Two white stars:** Positioned on the black band

### üîß Technical Skills You'll Use:
- Drawing triangular polygons
- Calculating diagonal coordinates
- Precise color matching
- Layering shapes in correct order

### üí° Strategy:
1. Start with the colored triangular backgrounds
2. Add the diagonal black band
3. Add yellow border stripes
4. Place the white stars

Let's code the flag step by step!


```python
import math

print("üá∞üá≥ QUEST 1: THE NATIONAL FLAG")
print("Challenge: Recreate St. Kitts & Nevis flag with geometric precision")

# Flag dimensions (maintaining proper proportions)
flag_width = 600
flag_height = 400

# Official flag colors
FLAG_GREEN = (0, 128, 0)
FLAG_RED = (255, 0, 0) 
FLAG_BLACK = (0, 0, 0)
FLAG_YELLOW = (255, 255, 0)
FLAG_WHITE = (255, 255, 255)

# Create flag canvas
flag = Image.new('RGB', (flag_width, flag_height), FLAG_WHITE)
flag_draw = ImageDraw.Draw(flag)

print("Step 1: Drawing the triangular backgrounds...")

# Calculate triangle points
# Green triangle (upper left)
green_triangle_1 = [(0, 0), (flag_width//2, 0), (0, flag_height)]
flag_draw.polygon(green_triangle_1, fill=FLAG_GREEN)

# Green triangle (lower right)  
green_triangle_2 = [(flag_width, flag_height), (flag_width//2, flag_height), (flag_width, 0)]
flag_draw.polygon(green_triangle_2, fill=FLAG_GREEN)

# Red triangle (upper right)
red_triangle_1 = [(flag_width, 0), (flag_width//2, 0), (flag_width, flag_height)]
flag_draw.polygon(red_triangle_1, fill=FLAG_RED)

# Red triangle (lower left)
red_triangle_2 = [(0, flag_height), (flag_width//2, flag_height), (0, 0)]
flag_draw.polygon(red_triangle_2, fill=FLAG_RED)

print("Step 2: Adding the diagonal black band...")

# Black diagonal band (wider)
band_width = 80
# Calculate parallel lines for the band
offset = band_width // 2

# Black band points (from lower left to upper right)
black_band = [
    (0, flag_height - offset),
    (0, flag_height + offset), 
    (flag_width, offset),
    (flag_width, -offset)
]
flag_draw.polygon(black_band, fill=FLAG_BLACK)

print("Step 3: Adding yellow border stripes...")

# Yellow stripes (thinner bands on each side of black)
stripe_width = 15

# Upper yellow stripe
upper_stripe = [
    (0, flag_height - offset - stripe_width),
    (0, flag_height - offset),
    (flag_width, -offset),
    (flag_width, -offset - stripe_width)
]
flag_draw.polygon(upper_stripe, fill=FLAG_YELLOW)

# Lower yellow stripe  
lower_stripe = [
    (0, flag_height + offset),
    (0, flag_height + offset + stripe_width),
    (flag_width, offset + stripe_width),
    (flag_width, offset)
]
flag_draw.polygon(lower_stripe, fill=FLAG_YELLOW)

print("Step 4: Adding the two white stars...")

# Star drawing function
def draw_star(draw, center_x, center_y, size, color):
    """Draw a 5-pointed star"""
    points = []
    for i in range(10):  # 10 points for a star (5 outer, 5 inner)
        angle = math.pi * i / 5
        if i % 2 == 0:  # Outer points
            radius = size
        else:  # Inner points
            radius = size * 0.4
        x = center_x + radius * math.cos(angle - math.pi/2)
        y = center_y + radius * math.sin(angle - math.pi/2)
        points.append((x, y))
    draw.polygon(points, fill=color)

# Position stars on the black band
star1_x, star1_y = flag_width * 0.3, flag_height * 0.6
star2_x, star2_y = flag_width * 0.7, flag_height * 0.4

draw_star(flag_draw, star1_x, star1_y, 20, FLAG_WHITE)
draw_star(flag_draw, star2_x, star2_y, 20, FLAG_WHITE)

print("‚úÖ Flag complete!")

# Display the completed flag
plt.figure(figsize=(12, 8))
plt.imshow(flag)
plt.title("üá∞üá≥ St. Kitts and Nevis National Flag")
plt.axis('off')
plt.show()

# Save the flag
flag.save("st_kitts_nevis_flag.png")
print("üíæ Flag saved as 'st_kitts_nevis_flag.png'")

print("\nüéâ QUEST 1 COMPLETE!")
print("‚ú® You've successfully recreated the national flag using geometric shapes!")
```
