````python
# 🎼 Step 1: Our Musical Notes

print("🎵 Building 'Mary Had a Little Lamb' Step by Step")

# These are the notes we can use
C4 = 261  # Middle C
D4 = 294  # D  
E4 = 330  # E
F4 =349  # F
G4 = 392  # G

print("🎹 Our musical notes:")
print(f"   C4 = {C4} Hz")
print(f"   D4 = {D4} Hz") 
print(f"   E4 = {E4} Hz")
print(f"   F4 = {F4} Hz")
print(f"   G4 = {G4} Hz")

print("\n🎵 'Mary Had a Little Lamb' uses these notes:")
print("   Ma-ry had a lit-tle lamb")
print("   E  D  C   D  E   E   E")

print("\n✅ Notes ready! Next: Create each note...")
````

````python
# 🎼 Step 2: Make Each Note We Need

print("🎼 Step 2: Creating Each Individual Note")

# Make each note for our song
note_E = create_steel_pan_note(E4, duration_ms=500)  # 0.5 seconds
note_D = create_steel_pan_note(D4, duration_ms=500)  # 0.5 seconds  
note_C = create_steel_pan_note(C4, duration_ms=500)  # 0.5 seconds
note_E_long = create_steel_pan_note(E4, duration_ms=800)  # 0.8 seconds (longer)

print("🎵 Created notes:")
print("   note_E = E4 for 0.5 seconds")
print("   note_D = D4 for 0.5 seconds")
print("   note_C = C4 for 0.5 seconds") 
print("   note_E_long = E4 for 0.8 seconds")

print("\n🎧 Listen to each note:")
play_audio(note_E, "Note E")
play_audio(note_D, "Note D") 
play_audio(note_C, "Note C")
play_audio(note_E_long, "Note E (long)")

print("\n✅ Individual notes ready! Next: Put them in order...")
````

````python
# 🎼 Step 3: Put the Notes Together in Order

print("🎼 Step 3: Putting Notes in Order")
print("'Ma-ry had a lit-tle lamb' = E D C D E E E")

# Put the notes together in the right order
mary_melody = note_E + note_D + note_C + note_D + note_E + note_E + note_E_long

print("🎵 Combined the notes:")
print("   note_E + note_D + note_C + note_D + note_E + note_E + note_E_long")

print("\n🎧 Listen to the complete melody:")
play_audio(mary_melody, "🐑 Mary Had a Little Lamb - Complete")

print("\n✅ Basic melody complete! Next: Add harmony...")
````

````python
# 🎼 Step 4: Make Harmony Notes (Higher Notes)

print("🎼 Step 4: Creating Harmony Notes")
print("Harmony = notes that sound good with the melody")

# For each melody note, we'll add a note that's higher
# E goes with G, D goes with F, C goes with E

harmony_G = create_steel_pan_note(G4, duration_ms=500)  # G harmony for E
harmony_F = create_steel_pan_note(F4, duration_ms=500)  # F harmony for D  
harmony_E = create_steel_pan_note(E4, duration_ms=500)  # E harmony for C
harmony_G_long = create_steel_pan_note(G4, duration_ms=800)  # Long G

print("🎼 Created harmony notes:")
print("   harmony_G = G4 (goes with melody E)")
print("   harmony_F = F4 (goes with melody D)")
print("   harmony_E = E4 (goes with melody C)")

# Put harmony notes in the same order as melody
mary_harmony = harmony_G + harmony_F + harmony_E + harmony_F + harmony_G + harmony_G + harmony_G_long

print("\n🎧 Listen to just the harmony:")
play_audio(mary_harmony, "🎼 Mary Harmony Part")

print("\n✅ Harmony ready! Next: Play them together...")
````

````python
# 🎼 Step 5: Play Melody and Harmony at the Same Time

print("🎼 Step 5: Playing Melody and Harmony Together")

# Use .overlay() to play both at the same time
mary_with_harmony = mary_melody.overlay(mary_harmony)

print("🎵 Combined melody and harmony:")
print("   mary_melody.overlay(mary_harmony)")

print("\n🎧 Listen to melody + harmony together:")
play_audio(mary_with_harmony, "🐑🎼 Mary with Harmony")

print("\n🎵 Compare all three versions:")
print("1. Just melody:")
play_audio(mary_melody, "🐑 Just Melody")

print("\n2. Just harmony:")  
play_audio(mary_harmony, "🎼 Just Harmony")

print("\n3. Both together:")
play_audio(mary_with_harmony, "🐑🎼 Both Together")

print("\n✅ Complete! Next: Use different sounds...")
````

````python
# 🎼 Step 6: Make Different Instrument Sounds

print("🎼 Step 6: Using Different Instrument Sounds")

# Make melody notes with a flute sound (sine wave)
flute_E = Sine(E4).to_audio_segment(duration=500).apply_gain(-10)
flute_D = Sine(D4).to_audio_segment(duration=500).apply_gain(-10)
flute_C = Sine(C4).to_audio_segment(duration=500).apply_gain(-10)
flute_E_long = Sine(E4).to_audio_segment(duration=800).apply_gain(-10)

# Put flute notes together
mary_flute = flute_E + flute_D + flute_C + flute_D + flute_E + flute_E + flute_E_long

print("🎵 Created flute version using Sine waves")

# Make harmony with warm pad sound (triangle wave)
pad_G = Triangle(G4).to_audio_segment(duration=500).apply_gain(-12)
pad_F = Triangle(F4).to_audio_segment(duration=500).apply_gain(-12)
pad_E = Triangle(E4).to_audio_segment(duration=500).apply_gain(-12)
pad_G_long = Triangle(G4).to_audio_segment(duration=800).apply_gain(-12)

# Put pad notes together
mary_pad = pad_G + pad_F + pad_E + pad_F + pad_G + pad_G + pad_G_long

print("🎼 Created pad harmony using Triangle waves")

# Combine flute melody with pad harmony
mary_orchestrated = mary_flute.overlay(mary_pad)

print("\n🎧 Listen to different instrument sounds:")
print("1. Steel pan version:")
play_audio(mary_with_harmony, "🎵 Steel Pan Version")

print("\n2. Flute + Pad version:")
play_audio(mary_orchestrated, "🎵 Flute + Pad Version")

print("\n✅ Different instruments complete!")
````
